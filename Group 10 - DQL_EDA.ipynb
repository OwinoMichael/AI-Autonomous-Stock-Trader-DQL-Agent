# Stock Data Analysis for Deep Q-Learning Portfolio

This notebook explores four types of data analysis on stock market data, using the dataset `all_stocks_5yr.csv`. These include:


1. **Time Series Trend Analysis**  
2. **Volatility Analysis**  
3. **Volume Analysis**  
4. **Correlation Analysis**  


```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("all_stocks_5yr.csv")
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python

aapl_df = df[df['Name'] == 'AAPL']
plt.figure(figsize=(12, 6))
plt.plot(aapl_df['date'], aapl_df['close'], label='Close Price', color='blue')
plt.title('AAPL Stock Closing Price Trend (2013-2018)')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


googl_df = df[df['Name'] == 'GOOGL']
plt.figure(figsize=(12, 6))
plt.plot(googl_df['date'], googl_df['close'], label='Close Price', color='blue')
plt.title('GOOGL Stock Closing Price Trend (2013-2018)')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

amzn_df = df[df['Name'] == 'AMZN']
plt.figure(figsize=(12, 6))
plt.plot(amzn_df['date'], amzn_df['close'], label='Close Price', color='blue')
plt.title('AMZN Stock Closing Price Trend (2013-2018)')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

fb_df = df[df['Name'] == 'FB']
plt.figure(figsize=(12, 6))
plt.plot(fb_df['date'], fb_df['close'], label='Close Price', color='blue')
plt.title('AMZN Stock Closing Price Trend (2013-2018)')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

msft_df = df[df['Name'] == 'MSFT']
plt.figure(figsize=(12, 6))
plt.plot(msft_df['date'], msft_df['close'], label='Close Price', color='blue')
plt.title('MSFT Stock Closing Price Trend (2013-2018)')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

adbe_df = df[df['Name'] == 'ADBE']
plt.figure(figsize=(12, 6))
plt.plot(adbe_df['date'], adbe_df['close'], label='Close Price', color='blue')
plt.title('ADBE Stock Closing Price Trend (2013-2018)')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

nflx_df = df[df['Name'] == 'NFLX']
plt.figure(figsize=(12, 6))
plt.plot(nflx_df['date'], nflx_df['close'], label='Close Price', color='blue')
plt.title('NFLX Stock Closing Price Trend (2013-2018)')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## 4. Correlation Analysis

```python

selected_companies = ['AAPL', 'GOOG', 'MSFT', 'AMZN', 'FB']
subset_df = df[df['Name'].isin(selected_companies)]
pivot_df = subset_df.pivot(index='date', columns='Name', values='close')
correlation_matrix = pivot_df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Selected Tech Stocks')
plt.tight_layout()
plt.show()
```
